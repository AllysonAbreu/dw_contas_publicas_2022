# Import e tratamento dos dados

## Import de dados

```python
import pandas as pd
import chardet
```

```python
with open("02_dados_pre_tratados/dados_instituto_previdencia_2021.csv", "rb") as f:
    result = chardet.detect(f.read())

df_instituto_previdencia_2021 = pd.read_csv("02_dados_pre_tratados/dados_instituto_previdencia_2021.csv", sep=";", encoding=result['encoding'])
df_instituto_previdencia_2021
```

```python
df_sctrans_2021 = pd.read_csv("02_dados_pre_tratados/dados_sctrans_2021.csv", sep=";", encoding=result['encoding'])
df_sctrans_2021
```

```python
df_prefeitura_municipal_2021 = pd.read_csv("02_dados_pre_tratados/dados_prefeitura_municipal_2021.csv", sep=";", encoding=result['encoding'])
df_prefeitura_municipal_2021
```

## Tratamento dos DataFrames

```python
# Definindo tipos das colunas antes de ler o dataset
columns_types = {'RECEITA': str, 'DATA': str}
```

```python
columns_Titles = ['NUMERO', 'CODIGO', 'RECEITA', 'DATA', 'HISTORICO', 'CONTA BANCARIA', 'FORNECEDOR', 'VALOR']
```

```python
def tratar_valores(valor):
#     print(f'Tamanho da string com pontuação: {len(valor)}')
    posicao_virgula = valor.find(',')
    posicao_ponto = valor.find('.')
#     print(f'Posição da vírgula: {posicao_virgula}')
#     print(f'Posição do ponto: {posicao_ponto}')
    
    string = ''.join(filter(str.isalnum, valor))
#     print(f'Tamanho da string sem pontuação: {len(string)}')
#     print(string)
    nova_posicao_ponto = posicao_virgula - (len(valor) - len(string))

    if len(string) == len(valor):
        return valor
    
    elif posicao_ponto != -1:        
        string = string[0:nova_posicao_ponto+1] + '.' + string[nova_posicao_ponto+1:]
        if valor[0] == '-':
            string = '-' + string
            return string
        return string
    
    else:
        if valor[0] == '-':
            string = string[0:posicao_virgula-1] + '.' + string[posicao_virgula-1:]       
            string = '-' + string
            return string
        
        string = string[0:posicao_virgula] + '.' + string[posicao_virgula:]        
        return string
    
    '''
        Função para separar a string que carregava o valor do código e da descrição do que cada um significava
    '''    
def tratar_df(df, column, new_column):
    '''
        Há um padrão nas strings. O código numérico vai até a possição 11, o restante é o separador, hífen, que não foi armazeando,
        e a descrição que fora armazenada na nova coluna.    
    '''
    df[new_column] = df[column].str.slice(0,16)
    df[column] = df[column].str.slice(17,)
    
def retirar_pontos_codigo(codigo):
    return codigo.replace(".", "")
```

### Tratando dados

```python
tratar_df(df_instituto_previdencia_2021, 'RECEITA', 'CODIGO')
df_instituto_previdencia_2021 = df_instituto_previdencia_2021.reindex(columns=columns_Titles)
df_instituto_previdencia_2021['CODIGO'] = df_instituto_previdencia_2021['CODIGO'].apply(retirar_pontos_codigo)
df_instituto_previdencia_2021['VALOR'] = df_instituto_previdencia_2021['VALOR'].apply(tratar_valores)
df_instituto_previdencia_2021
```

```python
tratar_df(df_prefeitura_municipal_2021, 'RECEITA', 'CODIGO')
df_prefeitura_municipal_2021 = df_prefeitura_municipal_2021.reindex(columns=columns_Titles)
df_prefeitura_municipal_2021['CODIGO'] = df_prefeitura_municipal_2021['CODIGO'].apply(retirar_pontos_codigo)
df_prefeitura_municipal_2021['VALOR'] = df_prefeitura_municipal_2021['VALOR'].apply(tratar_valores)
df_prefeitura_municipal_2021
```

```python
tratar_df(df_sctrans_2021, 'RECEITA', 'CODIGO')
df_sctrans_2021 = df_sctrans_2021.reindex(columns=columns_Titles)
df_sctrans_2021['CODIGO'] = df_sctrans_2021['CODIGO'].apply(retirar_pontos_codigo)
df_sctrans_2021['VALOR'] = df_sctrans_2021['VALOR'].apply(tratar_valores)
df_sctrans_2021
```

## Salvando dados tratados

```python
df_instituto_previdencia_2021.to_csv('03_dados_tratados/dados_assistencia_social_tratados_2021.csv', index=False, sep=';')
df_prefeitura_municipal_2021.to_csv('03_dados_tratados/dados_prefeitura_tratados_2021.csv', index=False, sep=';')
df_sctrans_2021.to_csv('03_dados_tratados/dados_sctrans_tratados_2021.csv', index=False, sep=';')
```

## Juntando todos os dados tratados em um único documento

```python
df_prefeitura_municipal_2021['TIPO'] = 1
df_instituto_previdencia_2021['TIPO'] = 2
df_sctrans_2021['TIPO'] = 3
```

```python
df_prefeitura_municipal_2021.drop('CONTA BANCARIA',axis=1, inplace=True)
df_instituto_previdencia_2021.drop('CONTA BANCARIA',axis=1, inplace=True)
df_sctrans_2021.drop('CONTA BANCARIA',axis=1, inplace=True)
```

```python
df_dados_acumulados_2021 = pd.merge(df_prefeitura_municipal_2021, df_instituto_previdencia_2021, how='outer')
df_dados_acumulados_2021 = pd.merge(df_dados_acumulados_2021, df_sctrans_2021, how='outer')
df_dados_acumulados_2021
```

```python
df_dados_acumulados_2021.drop('NUMERO', axis=1, inplace=True)
df_dados_acumulados_2021.drop('HISTORICO', axis=1, inplace=True)
df_dados_acumulados_2021
```
