## Tratamento dos DataFrames

```python
# Definindo tipos das colunas antes de ler o dataset
columns_types = {'Receita': str, 'Data': str}
```

```python
columns_Titles = [' Número Guia', 'Código', 'Receita', 'Data', 'Nome', 'CPF/CNPJ', 'Valor Previsto', 'Valor Lançado', 'Valor Arrecadado']
```

```python
def tratar_valores(valor):
#     print(f'Tamanho da string com pontuação: {len(valor)}')
    posicao_virgula = valor.find(',')
    posicao_ponto = valor.find('.')
#     print(f'Posição da vírgula: {posicao_virgula}')
#     print(f'Posição do ponto: {posicao_ponto}')
    
    string = ''.join(filter(str.isalnum, valor))
#     print(f'Tamanho da string sem pontuação: {len(string)}')
#     print(string)
    nova_posicao_ponto = posicao_virgula - (len(valor) - len(string))

    if len(string) == len(valor):
        return valor
    
    elif posicao_ponto != -1:        
        string = string[0:nova_posicao_ponto+1] + '.' + string[nova_posicao_ponto+1:]
        if valor[0] == '-':
            string = '-' + string
            return string
        return string
    
    else:
        if valor[0] == '-':
            string = string[0:posicao_virgula-1] + '.' + string[posicao_virgula-1:]       
            string = '-' + string
            return string
        
        string = string[0:posicao_virgula] + '.' + string[posicao_virgula:]        
        return string
    
    '''
        Função para separar a string que carregava o valor do código e da descrição do que cada um significava
    '''    
def tratar_df(df, column, new_column):
    '''
        Há um padrão nas strings. O código numérico vai até a possição 11, o restante é o separador, hífen, que não foi armazeando,
        e a descrição que fora armazenada na nova coluna.    
    '''
    df[new_column] = df[column].str.slice(0,10)
    df[column] = df[column].str.slice(11,) 
```

### Tratando dados

```python
tratar_df(df_instituto_previdencia_assistencia_social_2015, 'Receita', 'Código')
df_instituto_previdencia_assistencia_social_2015 = df_instituto_previdencia_assistencia_social_2015.reindex(columns=columns_Titles)
df_instituto_previdencia_assistencia_social_2015['Valor Previsto'] = df_instituto_previdencia_assistencia_social_2015['Valor Previsto'].apply(tratar_valores)
df_instituto_previdencia_assistencia_social_2015['Valor Arrecadado'] = df_instituto_previdencia_assistencia_social_2015['Valor Arrecadado'].apply(tratar_valores)
df_instituto_previdencia_assistencia_social_2015
```

```python
tratar_df(df_prefeitura_2015, 'Receita', 'Código')
df_prefeitura_2015 = df_prefeitura_2015.reindex(columns=columns_Titles)
df_prefeitura_2015['Valor Previsto'] = df_prefeitura_2015['Valor Previsto'].apply(tratar_valores)
df_prefeitura_2015['Valor Arrecadado'] = df_prefeitura_2015['Valor Arrecadado'].apply(tratar_valores)
df_prefeitura_2015
```

```python
tratar_df(df_sctrans_2015, 'Receita', 'Código')
df_sctrans_2015 = df_sctrans_2015.reindex(columns=columns_Titles)
df_sctrans_2015['Valor Previsto'] = df_sctrans_2015['Valor Previsto'].apply(tratar_valores)
df_sctrans_2015['Valor Arrecadado'] = df_sctrans_2015['Valor Arrecadado'].apply(tratar_valores)
df_sctrans_2015
```
