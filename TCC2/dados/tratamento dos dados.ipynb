```python
import pandas as pd
```

# Importando dados originais das Receitas

```python
df_receitas_2013 = pd.read_csv('dados_prefeitura/receitas_pm_cz/2013/01_dados originais/receita_tratado_2013.csv', sep=';')
df_receitas_2014 = pd.read_csv('dados_prefeitura/receitas_pm_cz/2014/01_dados originais/receita_tratado_2014.csv', sep=';')
df_receitas_2015 = pd.read_csv('dados_prefeitura/receitas_pm_cz/2015/01_dados originais/receita_tratado_2015.csv', sep=';')
df_receitas_2016 = pd.read_csv('dados_prefeitura/receitas_pm_cz/2016/01_dados originais/receita_tratado_2016.csv', sep=';')
df_receitas_2017 = pd.read_csv('dados_prefeitura/receitas_pm_cz/2017/01_dados originais/receita_tratado_2017.csv', sep=';')
df_receitas_2018 = pd.read_csv('dados_prefeitura/receitas_pm_cz/2018/01_dados originais/receita_tratado_2018.csv', sep=';')
df_receitas_2019 = pd.read_csv('dados_prefeitura/receitas_pm_cz/2019/01_dados originais/receita_tratado_2019.csv', sep=';')
df_receitas_2020 = pd.read_csv('dados_prefeitura/receitas_pm_cz/2020/01_dados originais/receita_tratado_2020.csv', sep=';')
```

```python
df_receitas_geral = pd.merge(df_receitas_2013,df_receitas_2014, how='outer')
df_receitas_geral = pd.merge(df_receitas_geral,df_receitas_2015, how='outer')
df_receitas_geral = pd.merge(df_receitas_geral,df_receitas_2016, how='outer')
df_receitas_geral = pd.merge(df_receitas_geral,df_receitas_2017, how='outer')
df_receitas_geral = pd.merge(df_receitas_geral,df_receitas_2018, how='outer')
df_receitas_geral = pd.merge(df_receitas_geral,df_receitas_2019, how='outer')
df_receitas_geral = pd.merge(df_receitas_geral,df_receitas_2020, how='outer')
df_receitas_geral
```

```python
df_contribuintes_geral = pd.DataFrame(df_receitas_geral[['CPF/CNPJ','Nome']])
df_contribuintes_geral
```

```python
df_contribuintes_geral.drop_duplicates(subset=['CPF/CNPJ'],keep='first', inplace=True)
df_contribuintes_geral
```

```python
df_contribuintes_geral.dropna(inplace=True)
```
