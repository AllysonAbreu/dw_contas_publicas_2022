## Tratamento dos DataFrames

```python
# Definindo tipos das colunas antes de ler o dataset
columns_types = {'VALOR':float,
                'LIQUIDADO':float,
                'PAGO':float,
                'SALDO':float}
```

```python
columns_Titles = ['NUMERO', 'DATA','CODIGO ORGAO', 'ORGAO', 'CODIGO DESPESA', 'DESPESA', 'DATA INCLUSAO', 'VALOR', 'LIQUIDADO', 'PAGO', 'SALDO']
```

```python
'''
Função para tratamento dos valores monetários

'''

def tratar_valores(valor):
    
    '''
        Primeiramente, procura-se a localização do ponto, que separava os milhares, e a vírgula que definia os centavos de reais.
    '''
    posicao_virgula = valor.find(',')
    posicao_ponto = valor.find('.') 
    
    '''
        Após o armazenamento das posições de cada separador, armazenava o valor em uma nova string, sem os pontos e vírgulas, 
        e encontrava a nova posição do separador, agora um ponto.
    '''
    string = ''.join(filter(str.isalnum, valor))
    nova_posicao_ponto = posicao_virgula - (len(valor) - len(string))

    '''
        Por fim, após verificar qual condição a nova string se encaixava, retornava-se a nova string com o devido tratamento.
    '''
    if len(string) == len(valor):
        return valor
    
    elif posicao_ponto != -1:        
        string = string[0:nova_posicao_ponto+1] + '.' + string[nova_posicao_ponto+1:]
        return string
    
    else:
        string = string[0:posicao_virgula] + '.' + string[posicao_virgula:]
        return string
    
    '''
        Função para separar a string que carregava o valor do código e da descrição do que cada um significava
    '''    
def tratar_df(df, column, new_column):
    '''
        Há um padrão nas strings. O código numérico vai até a possição 11, o restante é o separador, hífen, que não foi armazeando,
        e a descrição que fora armazenada na nova coluna.    
    '''
    if(column == 'ORGAO'):
        df[new_column] = df[column].str.slice(0,5)
        df[column] = df[column].str.slice(7,)
    else:
        df[new_column] = df[column].str.slice(0,7)
        df[column] = df[column].str.slice(8,)
```

### Tratando dados

```python
tratar_df(df_camara_2022, 'ORGAO', 'CODIGO ORGAO')
tratar_df(df_camara_2022, 'DESPESA', 'CODIGO DESPESA')
df_camara_2022 = df_camara_2022.reindex(columns=columns_Titles)

df_camara_2022['VALOR'] = df_camara_2022['VALOR'].apply(tratar_valores)
df_camara_2022['LIQUIDADO'] = df_camara_2022['LIQUIDADO'].apply(tratar_valores)
df_camara_2022['PAGO'] = df_camara_2022['PAGO'].apply(tratar_valores)
df_camara_2022['SALDO'] = df_camara_2022['SALDO'].apply(tratar_valores)

df_camara_2022
```
