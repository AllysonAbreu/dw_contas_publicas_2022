## Tratamento dos DataFrames

```python
# Definindo tipos das colunas antes de ler o dataset
columns_types = {'Empenhado(R$)':float,
                'Liquidado(R$)':float,
                'Pago(R$)':float,
                'Saldo(R$)':float}
```

```python
'''
Função para tratamento dos valores monetários

'''

def tratar_valores(valor):
    
    '''
        Primeiramente, procura-se a localização do ponto, que separava os milhares, e a vírgula que definia os centavos de reais.
    '''
    posicao_virgula = valor.find(',')
    posicao_ponto = valor.find('.') 
    
    '''
        Após o armazenamento das posições de cada separador, armazenava o valor em uma nova string, sem os pontos e vírgulas, 
        e encontrava a nova posição do separador, agora um ponto.
    '''
    string = ''.join(filter(str.isalnum, valor))
    nova_posicao_ponto = posicao_virgula - (len(valor) - len(string))

    '''
        Por fim, após verificar qual condição a nova string se encaixava, retornava-se a nova string com o devido tratamento.
    '''
    if len(string) == len(valor):
        return valor
    
    elif posicao_ponto != -1:        
        string = string[0:nova_posicao_ponto+1] + '.' + string[nova_posicao_ponto+1:]
        return string
    
    else:
        string = string[0:posicao_virgula] + '.' + string[posicao_virgula:]
        return string
```

### Tratando dados

```python
df_camara_2014['Empenhado(R$)'] = df_camara_2014['Empenhado(R$)'].apply(tratar_valores)
df_camara_2014['Pago(R$)'] = df_camara_2014['Pago(R$)'].apply(tratar_valores)
df_camara_2014['Liquidado(R$)'] = df_camara_2014['Liquidado(R$)'].apply(tratar_valores)

df_camara_2014
```

```python
df_instituto_previdencia_2014['Empenhado(R$)'] = df_instituto_previdencia_2014['Empenhado(R$)'].apply(tratar_valores)
df_instituto_previdencia_2014['Pago(R$)'] = df_instituto_previdencia_2014['Pago(R$)'].apply(tratar_valores)
df_instituto_previdencia_2014['Liquidado(R$)'] = df_instituto_previdencia_2014['Liquidado(R$)'].apply(tratar_valores)
df_instituto_previdencia_2014['Saldo(R$)'] = df_instituto_previdencia_2014['Saldo(R$)'].apply(tratar_valores)

df_instituto_previdencia_2014
```

```python
df_prefeitura_2014['Empenhado(R$)'] = df_prefeitura_2014['Empenhado(R$)'].apply(tratar_valores)
df_prefeitura_2014['Pago(R$)'] = df_prefeitura_2014['Pago(R$)'].apply(tratar_valores)
df_prefeitura_2014['Liquidado(R$)'] = df_prefeitura_2014['Liquidado(R$)'].apply(tratar_valores)
df_prefeitura_2014['Saldo(R$)'] = df_prefeitura_2014['Saldo(R$)'].apply(tratar_valores)

df_prefeitura_2014
```

```python
df_sctrans_2014['Empenhado(R$)'] = df_sctrans_2014['Empenhado(R$)'].apply(tratar_valores)
df_sctrans_2014['Pago(R$)'] = df_sctrans_2014['Pago(R$)'].apply(tratar_valores)
df_sctrans_2014['Liquidado(R$)'] = df_sctrans_2014['Liquidado(R$)'].apply(tratar_valores)
df_sctrans_2014['Saldo(R$)'] = df_sctrans_2014['Saldo(R$)'].apply(tratar_valores)

df_sctrans_2014
```
