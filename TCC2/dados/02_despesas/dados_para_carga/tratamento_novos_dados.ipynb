### Concatenando os valores que estão em células diferentes

```python
def concat_celulas(df):
    # concatenar as informações das células
    for i in range(0, len(df), 2):
        if i+1 < len(df):
            df.loc[i, 'TIPO/SUB-ELEMENTO'] = str(df.loc[i, 'TIPO/SUB-ELEMENTO']) + ' ' + str(df.loc[i+1, 'TIPO/SUB-ELEMENTO'])
            df.loc[i, 'FAVORECIDO'] = str(df.loc[i, 'FAVORECIDO']) + ' ' + str(df.loc[i+1, 'FAVORECIDO'])
            if not df.loc[i, 'LICITAÇÃO'] == 'Sem Licitação':
                df.loc[i, 'LICITAÇÃO'] = str(df.loc[i, 'LICITAÇÃO']) + '-' + str(df.loc[i+1, 'LICITAÇÃO'])
    # excluir linhas com Número igual a NaN
    df = df[df['NÚMERO'].notna()]
    # reindexar o dataframe
    df = df.reset_index(drop=True)
    # alterando o valor do conteúdo da coluna
    df['NÚMERO'] = df['NÚMERO'].astype('int64')
    
    return df
```

## Padronizando as colunas

```python
def extrair_valores_subelemento_e_documento(df):
    # cria as colunas SUBELEMENTO e DOCUMENTO com valores nulos
    df['SUBELEMENTO'] = pd.Series(dtype='object')
    df['DOCUMENTO'] = pd.Series(dtype='object')

    # regex para extrair o subelemento e o documento das células
    regex_subelemento = r'(?<=\s)\d+\s-\s[^\n]+'
    regex_documento = r'(?<=-\s)\d{2}\.\d{3}\.\d{3}/\d{4}-\d{2}'

    for i, row in df.iterrows():
        # extrai o subelemento da célula TIPO/SUB-ELEMENTO
        match_subelemento = re.search(regex_subelemento, str(row['TIPO/SUB-ELEMENTO']))
        if match_subelemento:
            subelemento = match_subelemento.group()
            df.at[i, 'SUBELEMENTO'] = subelemento

            # remove o subelemento da célula TIPO/SUB-ELEMENTO
            tipo_sem_subelemento = re.sub(regex_subelemento, '', str(row['TIPO/SUB-ELEMENTO']))
            df.at[i, 'TIPO'] = tipo_sem_subelemento.strip()

        # extrai o documento da célula FAVORECIDO
        match_documento = re.search(regex_documento, str(row['FAVORECIDO']))
        if match_documento:
            documento = match_documento.group()
            df.at[i, 'DOCUMENTO'] = documento

            # remove o número do documento da célula FAVORECIDO
            favorecido_sem_doc = re.sub(regex_documento, '', str(row['FAVORECIDO']))
            df.at[i, 'FAVORECIDO'] = favorecido_sem_doc.strip()

    return df
```

### Padronizando a coluna FUNCAO e SUBFUNCAO

```python
def tratar_coluna(df,column,new_column):
    # cria a coluna nova coluna com valores nulos
    df[new_column] = pd.Series(dtype='int64')

    # preenche os valores nulos da coluna em analise com uma string vazia
    df[column] = df[column].fillna('')

    for i, row in df.iterrows():
        # Extrai a parte numérica do início da string, e remove a parte numérica da string
        match = re.match(r'^(\d+)\s*-\s*(.*)$', row[column])
        if match:
            cod = match.group(1)
            string = match.group(2)
            string_sem_espaco = string.strip().lower()

            df.at[i, new_column] = int(cod)
            df.at[i, column] = unidecode.unidecode(string_sem_espaco)

    return df
```

```python
def limpar_string(valor):
    valor = unidecode(valor).lower()
    valor = re.sub(r'[^a-z0-9\s]', ' ', valor)
    valor = re.sub(r'\s+', ' ', valor)
    return valor
```

```python
def limpar_string_sem_numero(valor):
    valor = unidecode(valor).lower()
    valor = re.sub(r'[^a-z\s]', ' ', valor)
    valor = re.sub(r'\s+', ' ', valor)
    return valor
```

```python
def limpar_string_somente_numero(valor):
    if isinstance(valor, float):
        valor = str(valor)
    valor = valor.split()[0]
    valor = re.sub(r'[^0-9]', '', valor)
    return valor
```
