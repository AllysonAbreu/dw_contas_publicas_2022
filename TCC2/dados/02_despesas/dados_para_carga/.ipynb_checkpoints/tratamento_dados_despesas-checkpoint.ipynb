```python
dados_2020 = ['DATA','CPF/CNPJ','CONTRIBUINTE','VALOR','LIQUIDADO','PAGO','SALDO','TIPO']
dados_2022 = ['DATA','CONTRIBUINTE','VALOR','LIQUIDADO','PAGO','SALDO','TIPO']

df_dados_completos_2020 = pd.DataFrame(columns=dados_2020)
df_dados_completos_2022 = pd.DataFrame(columns=dados_2022)
```

```python
def merge_dataframes(lista_dataframes, colunas_sem_valores):
    
    df_final = colunas_sem_valores.copy() # cria uma cópia do DataFrame das colunas sem valores
    
    for df in lista_dataframes:
        df_final = pd.merge(df_final, df, on=list(df_final.columns), how='outer') # mescla o DataFrame atual com o DataFrame das colunas
        
    return df_final
```

```python
df_dados_completos_2020 = merge_dataframes(list_data_frames[:8],df_dados_completos_2020)

df_dados_completos_2020.drop('NUMERO', axis=1, inplace=True)
df_dados_completos_2020
```

```python
def tratar_coluna(df, coluna):
    
    # remove sinais de pontuação e símbolos
    table = str.maketrans('', '', string.punctuation)
    df[coluna] = df[coluna].apply(lambda x: str(x).translate(table))
    
    # remove acentuação gráfica
    df[coluna] = df[coluna].apply(lambda x: unidecode(x))
    
    # converte todos os caracteres para minúsculas
    df[coluna] = df[coluna].str.lower()
    
    return df
```

```python
def compare_columns(df_comparador, column_comparador, df_comparacao, column_comparacao, doc_column):
    # Cria cópias dos dataframes para evitar alterações indesejadas nos dados originais
    df1_copy = df_comparador.copy()
    df2_copy = df_comparacao.copy()

    # Cria uma nova coluna nos dataframes, caso ela ainda não exista
    if doc_column not in df1_copy.columns:
        df1_copy[doc_column] = pd.Series(dtype=str)
    if doc_column not in df2_copy.columns:
        df2_copy[doc_column] = pd.Series(dtype=str)

    # Remove valores numéricos e caracteres especiais das colunas de comparação
    df1_copy[column_comparador] = df1_copy[column_comparador].apply(lambda x: re.sub('[^A-Za-z ]+', '', x))
    df2_copy[column_comparacao] = df2_copy[column_comparacao].apply(lambda x: re.sub('[^A-Za-z ]+', '', x))

    # Cria as listas dos valores similares e não similares
    valores_similares = []
    valores_nao_similares = []

    # Percorre cada linha dos dataframes usando loops aninhados
    for index1, row1 in df1_copy.iterrows():
        is_similar = False
        for index2, row2 in df2_copy.iterrows():
            # Calcula a similaridade entre as duas strings usando a função `fuzz.token_sort_ratio()`
            similarity = fuzz.token_sort_ratio(row1[column_comparador], row2[column_comparacao])
            # Verifica se a similaridade é superior a 90%
            if similarity > 90:
                is_similar = True
                # Verifica se o número do documento já está presente em um dos dataframes
                if pd.isna(row1[doc_column]):
                    df1_copy.at[index1, doc_column] = row2[doc_column]
                elif pd.isna(row2[doc_column]):
                    df2_copy.at[index2, doc_column] = row1[doc_column]
                valores_similares.append(row2[column_comparacao])
                break
        if not is_similar:
            valores_nao_similares.append(row1[column_comparador])

    # Realiza o merge dos dataframes
    merged_df = pd.merge(df1_copy, df2_copy, how='outer')

    # Retorna os valores similares, não similares e o merge dos dataframes
    return valores_similares, valores_nao_similares, merged_df
```

```python
valores_similares, valores_nao_similares, df_juncao_contribuintes_gerais = compare_columns(df_contribuintes_receitas_2020,'CONTRIBUINTE',df_contribuintes_receitas_2022,'CONTRIBUINTE','CPF/CNPJ')
```

```python
def preencher_cpf_cnpj(df):
    # Cria uma cópia do dataframe para evitar a perda dos dados originais
    df_novo = df.copy()

    # Verifica se existem células vazias na coluna 'CPF/CNPJ'
    cnpj_vazio = df_novo['CPF/CNPJ'].isnull() | (df_novo['CPF/CNPJ'] == '')

    # Cria um array com valores únicos de CPF/CNPJ para evitar a criação de códigos repetidos
    cpf_cnpj_unicos = df_novo['CPF/CNPJ'].unique()

    # Itera pelas linhas do dataframe com CPF/CNPJ vazio
    for i, row in df_novo[cnpj_vazio].iterrows():
        # Gera um novo código único de até 14 dígitos
        novo_codigo = np.random.randint(10**13, 10**14, dtype=np.int64)

        # Verifica se o código gerado já existe no dataframe original
        while novo_codigo in cpf_cnpj_unicos:
            novo_codigo = np.random.randint(10**13, 10**14, dtype=np.int64)

        # Adiciona o novo código na célula vazia correspondente
        df_novo.loc[i, 'CPF/CNPJ'] = novo_codigo

        # Adiciona o novo código na lista de CPF/CNPJ únicos
        cpf_cnpj_unicos = np.append(cpf_cnpj_unicos, novo_codigo)

    return df_novo
```
