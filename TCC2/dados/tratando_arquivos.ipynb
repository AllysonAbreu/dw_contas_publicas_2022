### Dimensão tempo

```python
df_tempo_carga = pd.merge(df_2013,df_2014, how='outer')
df_tempo_carga = pd.merge(df_tempo_carga, df_2015, how = 'outer')
df_tempo_carga = pd.merge(df_tempo_carga, df_2016, how = 'outer')
df_tempo_carga = pd.merge(df_tempo_carga, df_2017, how = 'outer')
df_tempo_carga = pd.merge(df_tempo_carga, df_2018, how = 'outer')
df_tempo_carga = pd.merge(df_tempo_carga, df_2019, how = 'outer')
df_tempo_carga = pd.merge(df_tempo_carga, df_2020, how = 'outer')
df_tempo_carga
```

```python
df_tempo_carga.value_counts()
```

```python
df_tempo_carga.isnull().sum()
```

```python
"""

Criando os trimestres, semestres e renomeando os meses

"""

dic_trimestre = {1:'Primeiro Trimestre', 2:'Primeiro Trimestre',3:'Primeiro Trimestre',
               4:'Segundo Trimestre', 5:'Segundo Trimestre',6:'Segundo Trimestre',
               7:'Terceiro Trimestre',8:'Terceiro Trimestre',9:'Terceiro Trimestre',
               10:'Quarto Trimestre',11:'Quarto Trimestre',12:'Quarto Trimestre'}

dic_semestre = {1:'Primeiro Semestre', 2:'Primeiro Semestre',3:'Primeiro Semestre',
               4:'Primeiro Semestre', 5:'Primeiro Semestre',6:'Primeiro Semestre',
               7:'Segundo Semestre',8:'Segundo Semestre',9:'Segundo Semestre',
               10:'Segundo Semestre',11:'Segundo Semestre',12:'Segundo Semestre'}

dic_meses = {1:'janeiro',2:'fevereiro',3:'marco',
            4:'abril',5:'maio',6:'junho',
            7:'julho',8:'agosto',9:'setembro',
            10:'outubro',11:'novembro',12:'dezembro'}

dic_cod_meses = {'janeiro':1,'fevereiro':2,'marco':3,
            'abril':4,'maio':5,'junho':6,
            'julho':7,'agosto':8,'setembro':9,
            'outubro':10,'novembro':11,'dezembro':12}

dic_cod_trimestre = {'Primeiro Trimestre':1, 'Segundo Trimestre':2,
                    'Terceiro Trimestre':3,'Quarto Trimestre':4}
dic_cod_semestre = {'Primeiro Semestre':1, 'Segundo Semestre':2}
```

```python
df_tempo_carga['Mes'] = df_tempo_carga['Mes'].map(dic_meses)
```

```python
df_tempo_carga['Cod_Mes'] = df_tempo_carga['Mes'].map(dic_cod_meses)
```

```python
df_tempo_carga['Trimestre'] = df_tempo_carga['Mes'].map(dic_trimestre)
```

```python
df_tempo_carga['Cod_Trimestre'] = df_tempo_carga['Trimestre'].map(dic_cod_trimestre)
```

```python
df_tempo_carga['Semestre'] = df_tempo_carga['Mes'].map(dic_semestre)
```

```python
df_tempo_carga['Cod_Semestre'] = df_tempo_carga['Semestre'].map(dic_cod_semestre)
```

```python
df_tempo_carga
```

```python
df_tempo_carga = df_tempo_carga.sort_values(by=['Data'])
df_tempo_carga
```
